## Comparative Analysis
### pyhmc & pystan

```python
import numpy as np
import pandas as pd
import scipy.linalg as la
import seaborn as sns
import matplotlib.pyplot as plt
from pyhmc import hmc
```

## Pyhmc
### Example 1

```python
def logprob(theta):
    logp = 2 * theta**2 - theta**4
    grad = 4 * theta - 4 * theta**3
    return logp, grad

theta_0 = np.array([0])
```

```python
samples = hmc(logprob, x0=theta_0, n_samples=5000)
```

```python
plt1_hmc = sns.distplot(samples, kde = True, hist = False)
plt.suptitle("Density Plot Generated by pyhmc")
fig1_hmc = plt1_hmc.get_figure()
fig1_hmc.savefig("Fig1-ex1_pyhmc.png")
```

### Example 2

```python
from autograd import jacobian
import autograd.numpy as np
def log_prior(theta):
    return (-1/(2*10))*theta.T@theta

def log_like(theta, x):
    return np.log(0.5 * np.exp(-0.5*(theta[0]-x)**2) + 0.5* np.exp(-0.5*(theta[1]-x)**2))

def U(theta, x, n, batch_size):
    return -log_prior(theta) - (n/batch_size)*sum(log_like(theta, x))

gradU = jacobian(U, argnum = 0)

def logprob(theta):
    logp = np.sum(U(theta, x = x, n = n, batch_size = n))
    gradu = gradU(theta, x = x, n = n, batch_size = n).reshape((-1,))
    return logp, gradu

np.random.seed(663)
n = 200
theta = np.array([-3,3])
x = np.r_[np.random.normal(theta[0], 1, (n,1)),
              np.random.normal(theta[1], 1, (n,1))].reshape(-1,1)
```

```python
samples2 = hmc(logprob, x0=theta.reshape((-1)), n_samples=200, epsilon=0.01)
```

```python
plt2_hmc = sns.kdeplot(samples2[:,0], samples2[:,1])
plt.suptitle("Kernel Density Plot Generated by pyhmc")
fig2_hmc = plt2_hmc.get_figure()
fig2_hmc.savefig("Fig2-ex2_pyhmc.png")
```

## Pystan
### Example 1

```python
import pystan
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(663)
model_1 = '''
    functions {
          # log probability density function
          real ex1_lpdf (real theta) {return 2*theta^2 - theta^4;} #defined function in the paper
    }
    data {
    }
    parameters {
        real theta;
    }
    model {
        theta ~ ex1_lpdf();
    }
    
    '''

data_1 = {}
sm_stan1 = pystan.StanModel(model_code = model_1)
fit = sm_stan1.sampling(data = data_1, iter = 5000, chains = 4)
```

```python
plt1_stan = sns.kdeplot(fit["theta"])
plt.suptitle("Kernel Density Plot Generated by pystan")
fig1_stan = plt1_stan.get_figure()
fig1_stan.savefig('Fig1-ex1_pystan.png')
```
